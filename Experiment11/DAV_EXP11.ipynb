### Data Visualization libraries in Python

1. **Matplotlib:** Matplotlib is an open-source drawing library that supports various drawing types. You can generate plots, histograms, bar charts, and other types of charts with just a few lines of code.

>  Features:
* Flexible Plotting: Matplotlib provides a wide range of plotting options, allowing users to create various types of visualizations, including line plots, scatter plots, bar plots, histograms, and more.
* Customization: Users have fine-grained control over the appearance of plots, with the ability to customize colors, styles, labels, and annotations. This ensures that visualizations meet specific requirements.

> Applications:

2. **Seaborn:** Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

> Features:
* Statistical Data Visualization: Seaborn is built on top of Matplotlib and provides a high-level interface for creating informative and attractive statistical graphics.
* Ease of Use: Seaborn simplifies the creation of complex statistical visualizations with concise and expressive syntax, enabling users to generate informative plots with minimal code.
* Integration with Pandas: Seaborn seamlessly integrates with Pandas DataFrames, allowing for easy manipulation and visualization of tabular data.
* Time Series Visualization: Seaborn supports time series visualization, offering tools to plot time-dependent data with meaningful insights.


>Applications:
* Exploratory Data Analysis (EDA): Seaborn is widely used in data exploration to generate quick and informative visual summaries of datasets.
* Statistical Analysis: Researchers and data scientists leverage Seaborn for visualizing statistical relationships and conducting exploratory statistical analysis.
* Dashboard Creation: Seaborn is used to create interactive dashboards and visualizations for data-driven storytelling and presentations.


3. **Plotly:** Plotly library in Python is an open-source library that can be used for data visualization and understanding data simply and easily. Plotly supports various types of plots like line charts, scatter plots, histograms, box plots, etc.

> Features:
* Interactive Visualizations: Plotly specializes in creating interactive and dynamic visualizations, allowing users to explore data, zoom in, pan, and interact with plots in real-time.
* Web-Based Framework: Plotly is designed for web-based plotting and is suitable for creating interactive plots for web applications, dashboards, and online presentations.
* Dashboards and Applications: Plotly's Dash framework enables the creation of interactive web applications and dashboards directly from Python, making it a powerful tool for building data-driven applications.

> Applications:
* Data Exploration and Analysis: Plotly is widely used for exploring and analyzing data interactively, enabling users to gain insights into complex datasets.
* Dashboards and Reporting: Plotly is employed to create interactive dashboards and reports, providing a dynamic way to present and share data-driven insights.
* Scientific and Research Visualization: Researchers use Plotly to visualize scientific data, experimental results, and research findings, especially in fields such as physics, biology, and engineering.

4. **Geoplotlib:**
Geoplotlib is a powerful Python library for visualizing geographical data. It provides a simple and intuitive way to create maps and plot geographical data on them. Geoplotlib is built on top of matplotlib and numpy. It is designed to make it easy to create maps and plot data on them.

> Features:
* Geospatial Visualization: geoplotlib is a Python library specifically designed for geospatial data visualization, allowing users to create maps and visualize geographical patterns.
* Customizable Maps: geoplotlib provides tools to create customizable maps, allowing users to control map styles, colors, and annotations to tailor visualizations to specific requirements.
* Marker Plots: The library supports marker plots, enabling the visualization of point-based geospatial data with markers, labels, and custom symbols.
* Heatmaps: geoplotlib allows the creation of heatmaps, which are useful for visualizing spatial density and intensity of data.


>Applications:
* Geographical Data Analysis: geoplotlib is primarily used for analyzing and visualizing geographical data, making it a valuable tool for geospatial data scientists and researchers.
* Location-Based Services: The library finds applications in location-based services and applications, providing the ability to visualize and analyze location-based data.
* Environmental Studies: Researchers in environmental science use geoplotlib to visualize spatial data related to climate, pollution, and natural resources.





```r
import matplotlib.pyplot as plt
```

```r
# Line Plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, label='Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.legend()
plt.show()
```

```r
# Scatter Plot

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y, color='teal', label='Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot')
plt.legend()
plt.show()
```

```r
#Bar Chart

categories = ['Category A', 'Category B', 'Category C']
values = [10, 20, 15]

plt.bar(categories, values, color='limegreen', label='Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Chart')
plt.legend()
plt.show()
```

```r
#Histogram

import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, color='crimson', label='Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.legend()
plt.show()
```

```r
import seaborn as sns
```

```r
tips = sns.load_dataset("tips")

sns.histplot(tips['total_bill'], kde=True, color='palegreen')
plt.title('Distribution of Total Bill')
plt.show()
```

```r
tips = sns.load_dataset("tips")

sns.boxplot(x='day', y='total_bill', data=tips, palette='Set3')
plt.title('Box Plot of Total Bill by Day')
plt.show()
```

```r
flights = sns.load_dataset("flights")

flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Passenger Count Heatmap')
plt.show()
```

```r
# ggplot

library(ggplot2)

ggplot(mtcars, aes(x = mpg, y = disp)) +
  geom_point(color = "#FF3333") +
  labs(title = "Scatter Plot of MPG vs Displacement",
       x = "MPG",
       y = "Displacement") +
  theme_minimal() +
    theme(
    text = element_text(size = 16),
    axis.title.x = element_text(size = 18),
    axis.title.y = element_text(size = 18),
    axis.text.x = element_text(size = 14),
    axis.text.y = element_text(size = 14),
    plot.title = element_text(size = 20)
  )
```

```r
ggplot(diamonds, aes(x = cut, fill = cut)) +
  geom_bar() +
  labs(title = "Bar Chart of Diamond Cuts",
       x = "Cut",
       y = "Count") +
  theme_minimal() +
    theme(
    text = element_text(size = 16),
    axis.title.x = element_text(size = 18),
    axis.title.y = element_text(size = 18),
    axis.text.x = element_text(size = 14),
    axis.text.y = element_text(size = 14),
    plot.title = element_text(size = 20)
  )
```

```r
ggplot(economics, aes(x = date, y = unemploy)) +
  geom_line(aes(color = "Unemployment")) +
  geom_line(aes(y = psavert, color = "Personal Saving")) +
  labs(title = "Line Plot of Unemployment and Personal Saving",
       x = "Date",
       y = "Value") +
  theme_minimal() +
    theme(
    text = element_text(size = 16),
    axis.title.x = element_text(size = 18),
    axis.title.y = element_text(size = 18),
    axis.text.x = element_text(size = 14),
    axis.text.y = element_text(size = 14),
    plot.title = element_text(size = 20)
  )
```

```r
#Lattice

library(lattice)

x <- rnorm(100)
y <- rnorm(100)

xyplot(y ~ x,
       main = "Scatter Plot",
       xlab = "X-axis",
       ylab = "Y-axis",
       pch = 16,
       col = "blue",
       xlim = c(-3, 3),
       ylim = c(-3, 3)
)
```

```r
data <- rnorm(100)

histogram(~ data,
          main = "Histogram of Sample Data",
          xlab = "Value",
          ylab = "Frequency",
          col = "skyblue",
          border = "black",
          breaks = 10
)
```
